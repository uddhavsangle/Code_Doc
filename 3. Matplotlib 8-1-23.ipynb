# Bar Plot

```python
import matplotlib.pyplot as plt
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]

plt.xlabel('language',fontsize=15,color='b')
plt.ylabel('number',fontsize=15,color='b')
plt.title('Graph2024',fontsize=20,color='g')

plt.bar(x,y,color=['r','g','m','y'],width=0.6,align='center',edgecolor='b',linewidth=2,linestyle='--',
        alpha=0.7,label="popularity")
plt.legend()
plt.show()
```

```python
import numpy as np
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]
z=[30,25,35,55]
width=0.3        # foe both column width so  use their
p=np.arange(len(x))# arrange function (len (x)-- as index no)
p1=[j+width for j in p]
plt.xlabel('language',fontsize=15,color='b')
plt.ylabel('number',fontsize=15,color='b')
plt.title('Graph2024',fontsize=20,color='g')

plt.bar(p,y,width,color='r',label='popularity')     # change x=p
plt.bar(p1,z,width,color='g',label='xcvb')            # change x= p --index no wise bar shows / p1 for side bar 

plt.xticks(p+width/2,x,rotation=30)
plt.legend(loc=5)         # location
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]
z=[30,25,35,55]

width=0.3        # foe both column width so  use their
p=np.arange(len(x))# arrange function (len (x)-- as index no)
p1=[j+width for j in p]

plt.xlabel('language',fontsize=15,color='b')
plt.ylabel('number',fontsize=15,color='b')
plt.title('Graph2024',fontsize=20,color='g')

plt.barh(p,y,width,color='r',label='popularity')     # change x=p
plt.barh(p1,z,width,color='g',label='male')            # change x= p --index no wise bar shows / p1 for side bar 
#plt.barh --for horizontal chart

plt.xticks(p+width/2,rotation=30)

plt.legend()         # location
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,7,9,5,8,8]
# c=['r','g','b','m','y','r']
colors=[20,70,35,50,60,70]
s=[500,90,60,300,400,80]
plt.scatter(x,y,c=colors,cmap="Accent",s=s,edgecolor="r",alpha=0.8)
t=plt.colorbar()
t.set_label('colorbar',fontsize=20,color='r',alpha=0.8)
plt.title('Scatter plot',fontsize=25,alpha=0.9)

plt.xlabel("day",fontsize=15,color='b')
plt.ylabel('no',fontsize=15,color='b')
plt.show()
```

```python
# rotate labels True/False

x=[20,40,30,10]
y=["java","python","c++","c"]
ex=[0.4,0.0,0.0,0.0] 
c=["m","g","r","b"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.0f%%",shadow=True,radius=0.9,labeldistance=1.1,startangle=90,
       textprops={"fontsize":15,'color':"lime"},counterclock=False,
        wedgeprops={"linewidth":1.5,"width":0.9,"edgecolor":"black"},rotatelabels=True)         # angle 
plt.title("piechart",color="b",fontsize=25)
plt.legend(loc=2)




plt.show()
```
