```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset("penguins")
df
```

```python
df.dropna()
```

```python
sns.barplot(x=df["island"],y=df["bill_length_mm"])
plt.show()
```

```python
#hue -- with respect 
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex")
plt.show()
```

```python
#order
order_1=["Torgersen","Biscoe island","Dream",]
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",order=order_1)
plt.show()
```

```python
order_1=["Torgersen","Biscoe island","Dream",]
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",order=order_1)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",hue_order=["female","male"],ci=90)
plt.show()
```

```python
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=df,hue="sex",ci=60,orient="h")
plt.show()
```

```python
sns.barplot(x=df["island"],y=df["bill_length_mm"],color="b",saturation=1,alpha=0.6)
plt.show()
```

```python
#palette=" accent"   use for new multi color
#saturation -- use for darkness low or high ( 0.1 to 1)
#errcolor -- error point color
#errwidth --- error point width
#capsize-- cap size 0 to 1

sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",palette="afmhot",saturation=0.1,errcolor="r",errwidth=2,
            capsize=0.5,)
plt.show()
```

```python
#dodge= True/False
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",errcolor="r",errwidth=2,
            capsize=0.5,dodge=False,alpha=0.9)
plt.show()
```

```python
#alpha
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",alpha=0.5,)
plt.show()
```

```python
#grid
sns.set(style="darkgrid")
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",alpha=0.7,)
plt.show()
```

# Heat map

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df= np.linspace(1,10,20).reshape(4,5)
df
```

```python
#vmin/vmax -- range 
#cmap-- for default color selection
#annot= TRUE/False -- for show value

sns.heatmap(df,vmin=1,vmax=10,cmap="Greens",annot=True)
```

```python
x=df1.drop(columns = ["attnr"],axis=1).head(10)
x
```

```python
sns.heatmap(x,vmin=1,vmax=10,annot=True)
```

```python
df2= np.linspace(1,10,10).reshape(2,5)
df2
```

```python
sns.heatmap(df2,vmin=1,vmax=10,cmap="PuOr",annot=True)
plt.show()
```

```python
ar=np.array([["a0","a1","a2","a3","a4"],["b0","b1","b2","b3","b4"]])
ar
```

```python
#annot=ar
#fmt=string
#annot_kws= for fontsize,color etc
#linewidth
y={"fontsize":12,"color":"r"}
sns.heatmap(df2,vmin=1,vmax=10,cmap="PuOr",annot=ar,fmt="s",annot_kws=y,linewidth=6,linecolor="y",cbar=False,xticklabels=False)
plt.show()
```

```python


v=sns.heatmap(df2,vmin=1,vmax=10,cmap="PuOr",annot=ar,fmt="s",annot_kws=y,
              linewidth=6,linecolor="y",cbar=False,xticklabels=False)
v.set(xlabel="python",ylabel="java")
sns.set(font_scale=8)
plt.show()
```
