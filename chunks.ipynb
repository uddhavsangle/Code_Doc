```python
import csv
import pandas as pd

def read_csv_in_chunks(file_path, chunk_size):
    with open(file_path, 'r', newline='') as csvfile:
        reader = csv.reader(csvfile)
        header = next(reader)  # Read the header
        while True:
            try:
                chunk = [next(reader) for _ in range(chunk_size)]
            except StopIteration:
                break
            if not chunk:
                break
            yield header, chunk

file_path = r"D:\notes\files\ccc.csv"
chunk_size = 2

# Iterate over chunks and print each chunk as a DataFrame
for header, chunk in read_csv_in_chunks(file_path, chunk_size):
    df = pd.DataFrame(chunk, columns=header)
    print(df)
```

```python
import csv
import pandas as pd
import numpy as np

def read_csv_in_chunks(file_path, chunk_size):
    with open(file_path, 'r', newline='') as csvfile:
        reader = csv.reader(csvfile)
        header = next(reader)  # Read the header
        while True:
            try:
                chunk = [next(reader) for _ in range(chunk_size)]
            except StopIteration:
                break
            if not chunk:
                break
            # Remove rows that are entirely null
            chunk = [row for row in chunk if not all(map(lambda x: x.strip() == '', row))]
            yield header, chunk

file_path = r"D:\notes\files\ccc.csv"
chunk_size = 2

# Iterate over chunks and print each chunk as a DataFrame
for header, chunk in read_csv_in_chunks(file_path, chunk_size):
    df = pd.DataFrame(chunk, columns=header)
    print(df)
```

```python
import csv
import pandas as pd
import numpy as np

def read_csv_in_chunks(file_path, chunk_size):
    with open(file_path, 'r', newline='') as csvfile:
        reader = csv.reader(csvfile)
        header = next(reader) 
        data = [row for row in reader if any(row)]  
        for i in range(0, len(data), chunk_size):
            yield header, data[i:i + chunk_size]

file_path = r"D:\notes\files\ccc.csv"
chunk_size = 2

for header, chunk in read_csv_in_chunks(file_path, chunk_size):
    df = pd.DataFrame(chunk, columns=header)
    print(df)
```
