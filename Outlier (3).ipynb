# z- score Method

## formula for calulate z value
## z=(x-Xmean)/std

```python
import numpy as np
```

```python
a=np.array([200,2,9,3,8,7,8,5,4,6,500])
```

```python
a
```

```python
mean=np.mean(a)
std=np.std(a)
print(mean)
print(std)

th=0.5
```

```python
outlier=[]                                           # z_score > th      ----outlier     z_score < th   ----data
valid_data=[]
for x in a:
    z_score=(x-mean)/std

    if np.abs(z_score)>th:
        outlier.append(x)
    else:
        valid_data.append(x)
        
    
```

```python
outlier
```

```python
valid_data
```

```python
z_score
```
